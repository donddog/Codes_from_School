```python
import json
import pandas as pd
import glob
import os
from collections import defaultdict
```

```python
input_file = 'train'

json_data = glob.glob(os.path.join(input_file, 'dialogues_*'))

json_data
```

```python
new_dialogues = defaultdict(list)
         
for i in range(0,len(json_data)):
    with open(json_data[i], "r") as json_file:
        dialogues = json.load(json_file)
        for j in range(0,len(dialogues)):
            for k in range(0,len(dialogues[j]['turns'])):                
                for l in range(len(dialogues[j]['turns'][k]["frames"][0]["actions"])):
                    new_dialogues['dialogue_id'].append(dialogues[j]["dialogue_id"])
                    new_dialogues['services'].append(dialogues[j]["services"])
                    
                    new_dialogues['act_slot'].append({dialogues[j]['turns'][k]["frames"][0]["actions"][l]['act']: [dialogues[j]['turns'][k]["frames"][0]["actions"][l]['slot']]})
                    
#                     new_dialogues['act_slot'] = []
#                     if l == 0:
#                         new_dialogues['act_slot'] += dialogues[j]['turns'][k]["frames"][0]["actions"][l]['act']
#                         list(new_dialogues['act_slot'][l]) += new_dialogues['act_slot'].extend([dialogues[j]['turns'][k]["frames"][0]["actions"][l]['slot']])
#                     else:
#                         list(new_dialogues['act_slot'][l]) += new_dialogues['act_slot'].extend([dialogues[j]['turns'][k]["frames"][0]["actions"][l]['slot']])
                        
#                     new_dialogues['act_slot'].append([dialogues[j]['turns'][k]["frames"][0]["actions"][l]['slot']])
#                     set(new_dialogues['act_slot'])

                    if dialogues[j]['turns'][k]["speaker"] == 'USER':
                        new_dialogues['active_intent'].append(dialogues[j]['turns'][k]["frames"][0]["state"]['active_intent'])
                    else:
                        new_dialogues['active_intent'].append(None)

                    if dialogues[j]['turns'][k]["speaker"] == 'USER':
                        new_dialogues['requested_slots'].append(dialogues[j]['turns'][k]["frames"][0]["state"]['requested_slots'])
                    else:
                        new_dialogues['requested_slots'].append(None)

                    new_dialogues['speaker'].append(dialogues[j]['turns'][k]["speaker"])
                    new_dialogues['utterance'].append(dialogues[j]['turns'][k]["utterance"])
                    
                    new_dialogues['turn_count'].append(int(k+1))
```

```python
new_df = pd.DataFrame(new_dialogues)#, columns=['dialogue_id', 'services', 'act_slot', 'active_intent','requested_slots', 'speaker', 'utterance'])
```

```python
if not os.path.exists('changing_SGD_datasets.csv'):
    new_df.to_csv('changing_SGD_datasets.csv', index=False, mode='w', encoding='utf-8-sig')
else:
    new_df.to_csv('changing_SGD_datasets.csv', index=False, mode='a', encoding='utf-8-sig', header=False)
```
